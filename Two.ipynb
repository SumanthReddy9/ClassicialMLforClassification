```python
import pandas as pd
```

```python
data = pd.read_csv("Tillit_Data_Science_Tech_Test - data.csv")
```

```python
data.head()
```

```python
for col in data.columns:
    print("---------------------------------------------")
    print(data[col].value_counts())
```

```python
null_data = dict(data.isnull().sum())
```

```python
missing_data = data.isnull().sum()/len(data)

missing_data = missing_data[missing_data > 0]

missing_data.sort_values(inplace  = True)

missing_data
```

```python
missing_data = missing_data.to_frame()

missing_data.columns = ['Count']

missing_data.index.names = ['Names']

missing_data["Names"] = missing_data.index


missing_data
```

```python
missing_data.columns
```

```python
import seaborn as sns

sns.set(style = "darkgrid", color_codes = True, rc={'figure.figsize':(100,40)})

sns.barplot(x = "Names", y = "Count", data = missing_data)
```

```python
drop_cols = missing_data[missing_data["Count"] > 0.5]["Names"].tolist()
```

```python
data_after_drop_cols = data.drop(drop_cols, axis = 1)
```

```python
data_after_drop_cols.drop_duplicates(subset="borrower_id", inplace=True)
```

```python
data_after_drop_cols = data_after_drop_cols.drop(["borrower_id"], axis = 1)
```

```python
data_after_drop_cols.head()
```

```python
missing_data = data_after_drop_cols.isnull().sum()/len(data_after_drop_cols)

missing_data = missing_data[missing_data > 0]

missing_data.sort_values(inplace  = True)

missing_data
```

```python
miss_dict = dict(missing_data)
```

```python
for x in list(miss_dict.keys()):
    if data_after_drop_cols[x].dtype == 'object':
        data_after_drop_cols[x] = data_after_drop_cols[x].fillna(list(dict(data_after_drop_cols[x].value_counts()).keys())[0])
    else:
        data_after_drop_cols[x] = data_after_drop_cols[x].fillna(data_after_drop_cols[x].mean())
```

```python
data_after_drop_cols["term"].value_counts()
```

```python
data_after_drop_cols["term"] = data_after_drop_cols["term"].str.split(' ').str[0]
```

```python
data_after_drop_cols["term"] = pd.to_numeric(data_after_drop_cols["term"])
```

```python
data_after_drop_cols.head()
```

```python
from sklearn import preprocessing

for x in data_after_drop_cols.columns:
    if data_after_drop_cols[x].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(data_after_drop_cols[x].values)
        data_after_drop_cols[x] = lbl.transform(data_after_drop_cols[x].values)
```

```python
data_after_drop_cols.describe()
```

```python
data_after_drop_cols["outcome"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_after_drop_cols.drop(["outcome"], axis = 1), data_after_drop_cols["outcome"])
```

```python
print(len(x_train), len(y_train), len(x_test), len(y_test))
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def plot_distribution(dataset, cols=5, width=20, height=15, hspace=0.2, wspace=0.5):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(float(dataset.shape[1]) / cols)
    for i, column in enumerate(dataset.columns):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.set_title(column)
        if dataset.dtypes[column] == np.object:
            g = sns.countplot(y=column, data=dataset)
            substrings = [s.get_text()[:18] for s in g.get_yticklabels()]
            g.set(yticklabels=substrings)
            plt.xticks(rotation=25)
        else:
            g = sns.distplot(dataset[column])
            plt.xticks(rotation=25)
    
plot_distribution(x_train, cols=2, width=20, height=40, hspace=0.75, wspace=0.5)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_metrics(y_true, y_pred):
    acc_score = accuracy_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred)
    print('Accuracy:', acc_score)
    print('-------')
    print('Confusion matrix:')
    print(conf_matrix)
    print('-------')
    print('Classification report:')
    print(class_report)
```
